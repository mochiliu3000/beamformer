# Main function to simulate data and run algorithm

```python
# 1. Generate 10 random seeds
import os
import random
import numpy as np

dirs_to_create = ["data","log"]

for dir_to_create in dirs_to_create:
    if not os.path.exists(dir_to_create):
        os.makedirs(dir_to_create)
        print("Directory '" + dir_to_create + "' Created ")
    else:    
        print("Directory '" + dir_to_create + "' already exists")  

seed_list = [s for s in range(100,110)]
print("seeds: " + str(seed_list))

num_clocks_list = [6, 10, 20]
print("num_clocks: " + str(num_clock_list))
```

```python
# 2. For each seed, generate simulated data
for seed in seed_list:
    os.environ['seed'] = str(seed)
    %run ./Data\ Simulation-simplified.ipynb
print("=====================All simulated data is generated=====================")
```

```python
# 3. For each seed and num_clocks, run algorithm and write log
for seed in seed_list:
    for num_clocks in num_clocks_list:
        os.environ['seed'] = str(seed)
        os.environ['num_clocks'] = str(num_clocks)
        %run ./Report-clock-simplified.ipynb
print("=====================All algorithm runing is done=====================")
```
